```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
```

```python
data ={'names': ['steve', 'john,', 'richard', 'sarah', 'randy', 'michael', 'julie',],
       'age': [ 20,22,20,21,24,23,22],
       'gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
       'rank': [2,1,5,5,3,7,6]}

df = DataFrame(data)
df
```

```python
### Bar Plot using matplotlib
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
##### Bar Plot using seaborn 
plot=sns.barplot(df, x='names', y= 'age')
plot.set_title('Comparing Ages')
plt.show
```

Line Plot with matplotlib

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

Line Plot Seaborn 

```python
plot = sns.lineplot(df, x='names', y='age')
plot.set_title('Comparing Ages')
plt.show()
```

pie chart with matplotlib

```python
plt.pie(df['age'], labels=df['names'])
plt.title('Age Comparison')
plt.show
```

Seaborn Piechart

```python
colors =sns.color_palette('pastel') [0:5]
plt.pie(df['age'], labels=df['names'], colors=colors)
plt.show()
```

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,4
sb.set_style('whitegrid')
```

Creating statistical data graphics 

```python
address = r"C:\Users\HP\Desktop\swc-python\mtcars.csv"
cars = pd.read_csv(address)
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.displot(mpg) # plotting instagram using seaborn 
```

```python
cars.plot(kind='scatter', x='hp', y='mpg', c= ['darkgray'], s=150)
```

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True) # creating a scatter plto using seaborn
```

creating a pairplot using seaborn 

```python
sns.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset) # a pairplot from the subset of the cars dataset
plt.show
```

building boxplots 

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y = 'mpg', data = cars, hue='am', palette='hls') # using seaborn to generate boxplots
```
